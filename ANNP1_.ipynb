```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
y = sigmoid(x)
plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show() 

```

```python
x = np.linspace(-10,10,100)
def tanh(x):
    return np.tanh(x)
y = tanh(x)
plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```

```python
x = np.linspace(-10,10,100)
def relu(x):
    return np.maximum(0,x)
y = relu(x)
plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title("Relu Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.show()
```
